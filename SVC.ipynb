# Modelo hecho con Support Vector Machine

## Importamos las bibliotecas y subimos los valores

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
datos = pd.read_excel('/home/desarrollo10/Downloads/data.xlsx',header=3)
datos.head()
```

## Eliminamos la Columna "Unnamed : 0" 

```python
datos=datos.drop(["Unnamed: 0"], axis=1)
#datos.head()
datos.tail() #Cola 
```

## Clasificamos cada posible cada valor con los números del 1-6

```python
datos["Variable"] = np.where(datos["Variable"]=="Mixed", "1", datos["Variable"])
datos["Variable"] = np.where(datos["Variable"]=="sCha", "2", datos["Variable"])
datos["Variable"] = np.where(datos["Variable"]=="Infected", "3", datos["Variable"])
datos["Variable"] = np.where(datos["Variable"]=="ABC", "4", datos["Variable"])
datos["Variable"] = np.where(datos["Variable"]=="PPDC", "5", datos["Variable"])
datos["Variable"] = np.where(datos["Variable"]=="Ady", "6", datos["Variable"])
```

## Tomamos la variable Y con los valores de las Variables y X con los valores número de cada variable

```python
columns = datos.columns.tolist()   
Y = ["Variable"]

interest_colums=columns[2:]
X=interest_colums
datos[X].head()
```

## Creamos un arreglo con los valores de X y  de Y

```python

X=np.array(datos[columns[2:]])
Y=np.array(datos["Variable"])
```

## Creamos el modelo con Support Vector Clasification (SVC)

```python
from sklearn.svm import SVC

C = 10e4
svc = SVC(kernel="rbf", C=C, decision_function_shape="ovr").fit(X,Y)
```

## Dado los valores de X, predecimos los valores de Y y lo llamamos Ypred

```python
Ypred = svc.predict(X)
```

## Observamos el porcentaje de predicción entre Y y Ypred 

```python
from sklearn import metrics
metrics.accuracy_score( Ypred,Y)*100
```
