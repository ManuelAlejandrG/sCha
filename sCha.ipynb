## Cambiamos los valores de la columna variable:
* Mixed --> 1
* sCha --> 2
* Infected --> 3
* ABC --> 4
* PPDC --> 5
* Ady --> 6

```python
datos["Variable"] = np.where(datos["Variable"]=="Mixed", "1", datos["Variable"])
datos["Variable"] = np.where(datos["Variable"]=="sCha", "2", datos["Variable"])
datos["Variable"] = np.where(datos["Variable"]=="Infected", "3", datos["Variable"])
datos["Variable"] = np.where(datos["Variable"]=="ABC", "4", datos["Variable"])
datos["Variable"] = np.where(datos["Variable"]=="PPDC", "5", datos["Variable"])
datos["Variable"] = np.where(datos["Variable"]=="Ady", "6", datos["Variable"])

#datos["Variable"] = (datos["Variable"]=="sCha").astype(int)
datos
```

## Columnas para hacer el modelo 

```python

Y = ["Variable"]

interest_colums=columns[2:]
X=interest_colums

```

## Importamos el modelo y hacemos una selección para la validación


```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

X_train, X_test, y_train, y_test = train_test_split(datos[X], datos[Y], test_size=0.3, random_state=0)
logreg = LogisticRegression()
logreg.fit(X_train,y_train)
```

## Porcentaje de acierto del modelo 

```python
logreg.score(X_train,y_train)*100
```

## Coeficientes para cada valor 

```python
logreg.coef_
```

## Predicción para X test

```python
prediction = logreg.predict(X_test)
prediction
```

## Porcentaje de acierto entre X test y Y test

```python
from sklearn import metrics
metrics.accuracy_score( prediction,y_test)*100
```
